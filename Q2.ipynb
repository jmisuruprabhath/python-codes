```python
# Assume that there are 1000 people in each test and control groups
from scipy.stats import chi2_contingency

churn_test = 54
churn_control = 814
n_test = 2000
n_control = 2000

stat, p, dof, expected = chi2_contingency([[churn_test, (n_test - churn_test)], [churn_control, (n_control - churn_control)]])

print(f"Chi-squared statistic: {stat:.2f}")
print(f"P-value: {p:.2f}")
print(f"Degrees of freedom: {dof}")
print(f"Expected frequencies: {expected}")

alpha = 0.05

if p < alpha:
  print("There's a significant difference between the churn rates of test and control groups.")
else:
  print("There's no significant difference between the churn rates of test and control groups.")
```

```python
# Calculate churn rates for both groups.
# Measure the percentage reduction in churn.
# Perform a chi-squared test to check for statistical significance.
```

```python
import numpy as np

def churn_analysis(churn_test, n_test, churn_control, n_control):
  churn_rate_test = churn_test/n_test
  churn_rate_control = churn_control/n_control

  reduction = (churn_rate_control - churn_rate_test)/churn_rate_control

  stats, p, dof, expected = chi2_contingency([[churn_test, (n_test - churn_test)], [churn_control, (n_control - churn_control)]])

  standard_error_test = np.sqrt(churn_rate_test * (1 - churn_rate_test) / n_test)
  standard_error_control = np.sqrt(churn_rate_control * (1 - churn_rate_control) / n_control)

  z_score = 1.96

  margin_of_error_test = z_score * standard_error_test
  margin_of_error_control = z_score * standard_error_control

  lower_boundry_test = churn_rate_test - margin_of_error_test
  lower_boundry_control = churn_rate_control - margin_of_error_control
  upper_boundry_test = churn_rate_test + margin_of_error_test
  upper_boundry_control = churn_rate_control + margin_of_error_control

  return churn_rate_test, churn_rate_control, reduction, p, lower_boundry_test, lower_boundry_control, upper_boundry_test, upper_boundry_control
```
